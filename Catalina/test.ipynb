```python
from openai import OpenAI
```

```python
from dotenv import dotenv_values
```

```python
config = dotenv_values(".env")
```

```python
client = OpenAI(api_key=config["OPENAI_API_KEY"])
```

```python
prompt = """You are a compliance officer for one of the top Asset Manager companies in the world designed to output JSON. 
Write 1 compliance policy related to parental leave benefits. The policy should be 500 words long"""
```

```python
response = client.chat.completions.create(
    model = "gpt-3.5-turbo",
    response_format={ "type": "json_object"},
    messages =[{"role": "user", "content": prompt}]  
)
```

```python
print(response.choices[0].message.content)
```
